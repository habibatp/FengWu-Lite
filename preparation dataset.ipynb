```python
import os

def analyze_dataset(root_path):
    print("\n" + "=" * 60)
    print(f"📁 Dataset : {root_path}")
    print("=" * 60)

    total_files = 0
    total_size = 0
    extensions = {}

    for dirpath, dirnames, filenames in os.walk(root_path):
        level = dirpath.replace(root_path, "").count(os.sep)
        indent = " " * 4 * level
        print(f"{indent}📂 {os.path.basename(dirpath)}")

        subindent = " " * 4 * (level + 1)

        for f in filenames:
            file_path = os.path.join(dirpath, f)
            size = os.path.getsize(file_path)

            print(f"{subindent}📄 {f} ({size / (1024*1024):.2f} MB)")

            total_files += 1
            total_size += size

            ext = os.path.splitext(f)[1]
            extensions[ext] = extensions.get(ext, 0) + 1

    print("\n--- Résumé ---")
    print(f"📦 Nombre total de fichiers : {total_files}")
    print(f"💾 Taille totale : {total_size / (1024**3):.2f} GB")

    print("📊 Types de fichiers :")
    for ext, count in extensions.items():
        print(f"   {ext if ext else '[sans extension]'} : {count}")


# ==============================
# Chemins de tes datasets
# ==============================

surface_path = r"C:\Users\user\Desktop\Graphcast_Project\ERA5_surface_MA_ES_6h_1986_2026"
pressure_path = r"C:\Users\user\Desktop\Graphcast_Project\ERA5_pressure_full_parallel_resume_1986_2026"

# ==============================
# Analyse
# ==============================

analyze_dataset(surface_path)
analyze_dataset(pressure_path)
```

```python
import os
import zipfile
import tempfile
import numpy as np
import xarray as xr
from tqdm import tqdm
from scipy.ndimage import zoom

import shutil
import uuid

# ======================
# CONFIG
# ======================
year = 2011
month = 11

surface_path = r"C:\Users\user\Desktop\Graphcast_Project\ERA5_surface_MA_ES_6h_1986_2026"
pressure_path = r"C:\Users\user\Desktop\Graphcast_Project\ERA5_pressure_full_parallel_resume_1986_2026"
output_path = rf"C:\Users\user\Desktop\Graphcast_Project\ERA5_np_test_{year}_{month:02d}"

os.makedirs(output_path, exist_ok=True)

surface_vars = ["u10", "v10", "t2m", "msl"]
pressure_vars = ["z", "q", "u", "v", "t"]

target_lat = np.linspace(46.0, 20.0, 64)
target_lon = np.linspace(-15.0, 6.0, 64)

# ======================
# RESIZE CHANNEL-WISE
# ======================
def resize_channels(x):
    # x: [C, H, W]
    out = np.zeros((x.shape[0], 64, 64), dtype=np.float32)
    for c in range(x.shape[0]):
        out[c] = zoom(
            x[c],
            (64 / x.shape[-2], 64 / x.shape[-1]),
            order=1
        )
    return out

# ======================
# SURFACE ZIP LOADER

def load_surface_zip(surface_file):

    tmpdir = os.path.join("temp_extract", str(uuid.uuid4()))
    os.makedirs(tmpdir, exist_ok=True)

    # extraction
    with zipfile.ZipFile(surface_file, "r") as z:
        z.extractall(tmpdir)

    # trouver fichier instant
    files = os.listdir(tmpdir)
    instant_files = [f for f in files if "instant" in f]

    if len(instant_files) == 0:
        raise ValueError("No instant file found")

    instant_path = os.path.join(tmpdir, instant_files[0])

    # ouvrir dataset
    ds = xr.open_dataset(instant_path, engine="netcdf4")

    ds = ds.interp(latitude=target_lat, longitude=target_lon)

    arrays = []
    for var in surface_vars:
        arrays.append(ds[var].values.astype(np.float32))

    # 🔥 IMPORTANT : fermer avant suppression
    ds.close()

    # supprimer dossier proprement
    shutil.rmtree(tmpdir, ignore_errors=True)

    return np.stack(arrays, axis=1)

# ======================
# PRESSURE LOADER
# ======================
def load_pressure_month(year, month):
    year_month = f"{year}_{month:02d}"
    pressure_datasets = []

    for blk in range(1, 5):
        p_file = f"era5_pressure_{year_month}_blk{blk}.grib"
        p_path = os.path.join(pressure_path, p_file)

        if not os.path.exists(p_path):
            raise FileNotFoundError(p_path)

        ds = xr.open_dataset(
            p_path,
            engine="cfgrib",
            backend_kwargs={
                "filter_by_keys": {"typeOfLevel": "isobaricInhPa"},
                "indexpath": "",
                "errors": "ignore"
            }
        )

        pressure_datasets.append(ds)

    ds_pressure = xr.concat(
        pressure_datasets,
        dim="isobaricInhPa",
        coords="minimal",
        compat="override"
    )

    ds_pressure = ds_pressure.sortby("isobaricInhPa")
    ds_pressure = ds_pressure.interp(latitude=target_lat, longitude=target_lon)

    levels = ds_pressure["isobaricInhPa"].values
    print("Pressure levels:", len(levels), levels)

    if len(levels) != 37:
        raise ValueError(f"Expected 37 pressure levels, got {len(levels)}")

    arrays = []
    for var in pressure_vars:
        if var not in ds_pressure:
            raise ValueError(f"Missing pressure variable: {var}")

        arr = ds_pressure[var].values.astype(np.float32)  # [T, 37, 64, 64]

        if arr.shape[1] != 37:
            raise ValueError(f"{var} does not have 37 levels: {arr.shape}")

        arrays.append(arr)

    for ds in pressure_datasets:
        ds.close()

    # [T, 185, 64, 64]
    return np.concatenate(arrays, axis=1)

# ======================
# MAIN
# ======================
surface_file = os.path.join(
    surface_path,
    f"era5_surface_{year}_{month:02d}.nc"
)

print("Loading surface ZIP...")
surface_data = load_surface_zip(surface_file)
print("surface_data:", surface_data.shape)

print("Loading pressure GRIB...")
pressure_data = load_pressure_month(year, month)
print("pressure_data:", pressure_data.shape)

if surface_data.shape[0] != pressure_data.shape[0]:
    raise ValueError(
        f"Time mismatch: surface={surface_data.shape[0]}, pressure={pressure_data.shape[0]}"
    )

final_data = np.concatenate([surface_data, pressure_data], axis=1)
print("final_data:", final_data.shape)

if final_data.shape[1:] != (189, 64, 64):
    raise ValueError(f"Wrong final shape: {final_data.shape}")

# SAVE
for i in tqdm(range(final_data.shape[0]), desc="Saving .npy"):
    out_file = os.path.join(output_path, f"{year}_{month:02d}_{i:03d}.npy")
    np.save(out_file, final_data[i].astype(np.float32))

# CHECK
sample = np.load(os.path.join(output_path, f"{year}_{month:02d}_000.npy"))

print("\nCHECK:")
print("shape:", sample.shape)
print("dtype:", sample.dtype)
print("mean:", np.nanmean(sample))
print("std:", np.nanstd(sample))
print("nan count:", np.isnan(sample).sum())

if sample.shape == (189, 64, 64):
    print("✅ READY FOR FengWu-Lite TRAINING")
else:
    print("❌ NOT COMPATIBLE")
```

```python
import os
import zipfile
import shutil
import uuid
import traceback
import numpy as np
import xarray as xr
from tqdm import tqdm
from scipy.ndimage import zoom

# ======================
# PATHS
# ======================
surface_path = r"C:\Users\user\Desktop\Graphcast_Project\ERA5_surface_MA_ES_6h_1986_2026"
pressure_path = r"C:\Users\user\Desktop\Graphcast_Project\ERA5_pressure_full_parallel_resume_1986_2026"

output_path = r"C:\Users\user\Desktop\Graphcast_Project\ERA5_np_float32_1986_2026"
os.makedirs(output_path, exist_ok=True)

# ======================
# VARIABLES
# ======================
surface_vars = ["u10", "v10", "t2m", "msl"]
pressure_vars = ["z", "q", "u", "v", "t"]

target_lat = np.linspace(46.0, 20.0, 64)
target_lon = np.linspace(-15.0, 6.0, 64)

# ======================
# RESIZE CHANNEL-WISE
# ======================
def resize_channels(x):
    out = np.zeros((x.shape[0], 64, 64), dtype=np.float32)

    for c in range(x.shape[0]):
        out[c] = zoom(
            x[c],
            (64 / x.shape[-2], 64 / x.shape[-1]),
            order=1
        )

    return out

# ======================
# LOAD SURFACE
# ======================
def load_surface(surface_file):

    tmpdir = os.path.join("temp_extract", str(uuid.uuid4()))
    os.makedirs(tmpdir, exist_ok=True)

    try:
        with zipfile.ZipFile(surface_file, "r") as z:
            z.extractall(tmpdir)

        files = os.listdir(tmpdir)
        instant_files = [f for f in files if "instant" in f.lower()]

        if len(instant_files) == 0:
            raise ValueError(f"No instant file found inside {surface_file}")

        instant_file = instant_files[0]

        ds = xr.open_dataset(
            os.path.join(tmpdir, instant_file),
            engine="netcdf4"
        )

        ds = ds.interp(latitude=target_lat, longitude=target_lon)

        arrays = []

        for var in surface_vars:
            if var not in ds:
                raise ValueError(f"Surface variable missing: {var}")

            arrays.append(ds[var].values.astype(np.float32))

        surface_data = np.stack(arrays, axis=1)  # [T, 4, 64, 64]

        ds.close()

        return surface_data

    finally:
        shutil.rmtree(tmpdir, ignore_errors=True)

# ======================
# LOAD PRESSURE
# ======================
def load_pressure(year, month):

    ym = f"{year}_{month:02d}"
    datasets = []

    for blk in range(1, 5):

        p_file = f"era5_pressure_{ym}_blk{blk}.grib"
        p_path = os.path.join(pressure_path, p_file)

        if not os.path.exists(p_path):
            print(f"⚠️ Missing pressure block: {p_file}")
            continue

        try:
            ds = xr.open_dataset(
                p_path,
                engine="cfgrib",
                decode_timedelta=False,
                backend_kwargs={
                    "filter_by_keys": {"typeOfLevel": "isobaricInhPa"},
                    "indexpath": "",
                    "errors": "ignore"
                }
            )

            if len(ds.data_vars) == 0:
                print(f"⚠️ Empty pressure block: {p_file}")
                ds.close()
                continue

            # Garder seulement les variables utiles
            keep_vars = [v for v in pressure_vars if v in ds.data_vars]
            ds = ds[keep_vars]

            # Charger en mémoire pour éviter les erreurs lazy cfgrib
            ds = ds.load()

            datasets.append(ds)

        except Exception as e:
            print(f"⚠️ Cannot open {p_file}: {repr(e)}")
            continue

    if len(datasets) == 0:
        raise ValueError(f"No valid pressure files found for {ym}")

    ds_pressure = xr.concat(
        datasets,
        dim="isobaricInhPa",
        coords="minimal",
        compat="override"
    )

    ds_pressure = ds_pressure.sortby("isobaricInhPa")

    ds_pressure = ds_pressure.interp(
        latitude=target_lat,
        longitude=target_lon
    )

    arrays = []

    for var in pressure_vars:
        if var not in ds_pressure:
            raise ValueError(f"Pressure variable missing: {var} for {ym}")

        arrays.append(ds_pressure[var].values.astype(np.float32))

    pressure_data = np.concatenate(arrays, axis=1)

    for ds in datasets:
        ds.close()

    ds_pressure.close()

    return pressure_data

# ======================
# MAIN LOOP (1986 → 2026)
# ======================
for year in range(1986, 2027):

    for month in range(1, 13):

        if year == 2026 and month > 4:
            continue

        print(f"\n=== {year}-{month:02d} ===")

        surface_file = os.path.join(
            surface_path,
            f"era5_surface_{year}_{month:02d}.nc"
        )

        if not os.path.exists(surface_file):
            print(f"❌ Surface missing: {surface_file}")
            continue

        try:
            surface_data = load_surface(surface_file)
            pressure_data = load_pressure(year, month)

            print("Surface shape :", surface_data.shape)
            print("Pressure shape:", pressure_data.shape)

            if surface_data.shape[0] != pressure_data.shape[0]:
                print("❌ Time mismatch")
                continue

            final = np.concatenate([surface_data, pressure_data], axis=1)

            print("Final shape:", final.shape)

            if final.shape[1] != 189:
                print("❌ Wrong channel count")
                continue

            for i in tqdm(range(final.shape[0]), desc=f"Saving {year}-{month:02d}"):

                out_file = os.path.join(
                    output_path,
                    f"{year}_{month:02d}_{i:03d}.npy"
                )

                np.save(out_file, resize_channels(final[i]))

            print(f"✅ {year}-{month:02d} OK")

        except Exception as e:
            print(f"❌ ERROR {year}-{month:02d}: {repr(e)}")
            traceback.print_exc()

print("\n🎉 DATASET 1986–2026 READY")
```
